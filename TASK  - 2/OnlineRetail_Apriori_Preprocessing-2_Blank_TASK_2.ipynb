# Step 1: Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('OnlineRetail.csv', encoding = 'ISO-8859-1')
data
```

# Step 2: Drop Missing Invoices or Descriptions

```python
data2 = pd.read_csv('OnlineRetail.csv', encoding = 'ISO-8859-1')

# df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
# removes the row where the mising value (null). if how = all it removes the column when all data is null.
data_after_process = data2.dropna(how = 'any' ,subset = ["InvoiceNo", "Description"])
data_after_process
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
invoice_in_str = data_after_process['InvoiceNo'].astype(str)

temp = []  

for invoice in invoice_in_str :
    if(invoice.startswith('C')) :
        temp.append(True)
    else:
        temp.append(False)

print("Number of canceled invoices:", temp.count(True))

flag_list = []
for flag in temp :
    flag_list.append(not flag)

remaining_data = data_after_process[flag_list]
remaining_data
```

# Step 4: Keep Only Positive Quantities

```python
positive_data = remaining_data.query('Quantity > 0')

# data_after_process = positive_data.drop(positive_data.index)
# data_after_prccess
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
positive_data = positive_data.copy()  # used to solve the error
positive_data['Description'] = positive_data['Description'].str.lower()
positive_data['Description'] = positive_data['Description'].str.replace(' ', '')
positive_data


# or 
# positive_data.loc[:, 'Description'] = (
#     positive_data['Description']
#     .str.lower()
#     .str.replace(' ', '')
# )
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
data_uk_filtered = positive_data.query("Country == 'United Kingdom'")
data_uk_filtered
```

# Step 7: Remove Duplicates

```python
pure_data = data_uk_filtered.drop_duplicates()
pure_data
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
data_step_8 = pure_data.copy()
data_step_8['TotalPrice'] = data['Quantity'] * data['UnitPrice']
data_step_8
```

# Step 9 : Check for Top 10 Selling Products

```python
top_data = pure_data.groupby('Description')['Quantity'].sum().sort_values(ascending = False).head(10)
top_data

```

# Step 10 : Identify Most Active Customers

```python
top_products = pure_data.groupby('Description')['Quantity'].count()
top_products = top_products.sort_values(ascending=False)
top_products.head(10)


# top_products2 = pure_data.groupby('InvoiceNo').max()
# top_products
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
data['InvoiceMonth']=pd.to_datetime(data['InvoiceDate']).dt.month.value_counts().sort_index()
data
```
